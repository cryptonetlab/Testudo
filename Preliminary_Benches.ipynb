```python
import matplotlib.pyplot as plt
import numpy as np


### design 1
powers = [10, 12, 16, 20, 24, 26]
r1cs_cons = [pow(2,x) for x in powers]
groth_cons = [397681,467800,605769,744735,883976,952823]
```

```python
plt.plot(r1cs_cons, groth_cons)
plt.title("spartan r1cs input cons to groth16 cons")
plt.xlabel("r1cs cons")
plt.ylabel("groth16 cons")
```

```python
spartan_proving_time = [0.071,0.267,4.23,67.942,1088.789,4367.456]
```

```python
plt.plot(r1cs_cons, spartan_proving_time)
plt.title("spartan r1cs input cons to spartan proving")
plt.xlabel("r1cs cons")
plt.ylabel("proving times")
```

```python
groth16_proving_time = [8.824, 9.657, 14.009, 15.895, 17.715, 18.656]
plt.plot(r1cs_cons, groth16_proving_time)
plt.title("spartan r1cs input cons to groth16 proving time")
plt.xlabel("r1cs cons")
plt.ylabel("proving times")
```

```python
total_proving_time =[sum(x) for x in zip(spartan_proving_time, groth16_proving_time)]
total_proving_time
```

```python
plt.plot(r1cs_cons, total_proving_time)
plt.title("total proving time")
plt.xlabel("r1cs cons")
plt.ylabel("proving times")
```
